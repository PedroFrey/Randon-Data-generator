```python
!pip install Faker
!pip install faker_commerce
from faker import Faker
import pandas as pd
import random
import faker_commerce
fake = Faker(locale='pt-br')
fake.add_provider(faker_commerce.Provider)
```

```python
fake.__dir__()
```

```python
#Criar listas para armazenar os dados
id = []
job = []
name = []
date_of_birth = []
date_time = []
cryptocurrency_name = []
ecommerce_category = []
estado = []
salary = []

#Gerador de dados
Faker.seed(8) #Fixar qual conjunto de dados vamos obter
for _ in range(10):
  id.append(_)
  job.append(fake.job())
  name.append(fake.name())
  date_of_birth.append(fake.date_of_birth(minimum_age=18,maximum_age=65))
  date_time.append(fake.date_time())
  cryptocurrency_name.append(fake.cryptocurrency_name())
  ecommerce_category.append(fake.ecommerce_category())
  estado.append(fake.estado())
  salary.append(random.randrange(1,100) * 1_212) #Em termos de multiplos do salário mínimo
#Unificar as listas em um dicionário
dados = { 'id':id,
         'job':job,
         'name':name,
         'date_of_birth.':date_of_birth,
         'date_time':date_time,
         'cryptocurrency_name':cryptocurrency_name,
         'ecommerce_category':ecommerce_category,
         'estado':estado,
         'salary':salary
         }
```

```python
#Converter dicionário em uma tabela
df = pd.DataFrame(dados)
df
```
